### üîß Limpieza
```python
ventas['unidades'] = ventas['unidades'].fillna(0)
precios['precio'] = precios['precio'].fillna(precios['precio'].mean())
```

### üîó Merge
```python
df = ventas.merge(precios, on='producto', how='left')
df = df.merge(categorias, on='producto', how='left')
```

### üìà KPIs
```python
df['total'] = df['unidades'] * df['precio']
df['margen'] = df['total'] * 0.25
df['dia_semana'] = df['fecha'].dt.day_name()
```

### üìä Agrupaciones
```python
df.groupby('categoria')['total'].sum()
df.groupby('producto')['margen'].sum()
df.groupby('dia_semana')['total'].mean()
```

### üîÑ Pivot Table
```python
pd.pivot_table(df, values='total', index='categoria', columns='dia_semana', aggfunc='sum')
```

### üìâ Visualizaci√≥n
```python
df.groupby('categoria')['total'].sum().plot.bar()
```
