## 1Ô∏è‚É£ Preparaci√≥n del entorno


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

---
## 2Ô∏è‚É£ Dataset de ejemplo


````python
df = pd.DataFrame({
    'ventas': np.random.randint(10,100,50),
    'coste': np.random.randint(5,50,50),
    'categoria': np.random.choice(['A','B','C'], 50)
})
df.head()
````

---
## 3Ô∏è‚É£ Gr√°fico de l√≠neas (series num√©ricas)


````python
df['ventas'].plot(title='Ventas')
````

---
## 4Ô∏è‚É£ Histogramas
Para ver la distribuci√≥n de valores:

````python
df['ventas'].plot.hist(bins=10, alpha=0.7)
````

---
## 5Ô∏è‚É£ Gr√°fico de barras (agrupaciones)


````python
df.groupby('categoria')['ventas'].mean().plot.bar(title='Ventas promedio por categor√≠a')
````

---
## 6Ô∏è‚É£ Boxplot (distribuci√≥n estad√≠stica)
Perfecto para detectar outliers.

````python
df[['ventas','coste']].plot.box()
````

---
## 7Ô∏è‚É£ Series temporales

Creamos fechas y datos ficticios:

````python
fechas = pd.date_range('2024-01-01', periods=30)
ventas_dia = pd.Series(np.random.randint(50,200,30), index=fechas)
ventas_dia.plot(figsize=(10,4), title='Ventas por d√≠a')
````
