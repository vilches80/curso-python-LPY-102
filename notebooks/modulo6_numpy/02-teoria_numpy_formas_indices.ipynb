En este notebook aprender√°s a manipular la **estructura interna** de los arrays de NumPy:

- Cambiar formas (reshape)

## 1Ô∏è‚É£ Cambiar forma (`reshape`)

`reshape` *no copia datos*, solo cambia la vista (si es posible).

````python
import numpy as np

a = np.arange(12)
a_reshape = a.reshape(3,4)
a, a_reshape
````

El `-1` permite que NumPy deduzca la dimensi√≥n:

````python
a.reshape(2, -1)
````

---
## 2Ô∏è‚É£ Aplanar arrays

- `ravel()` devuelve una **vista** (si es posible)
- `flatten()` devuelve una **copia**

````python
x = np.array([[1,2,3],[4,5,6]])
x.ravel(), x.flatten()
````

`ravel()` modifica el original, `flatten()` no:

````python
r = x.ravel()
r[0] = 99
x
````

---
## 3Ô∏è‚É£ Indexaci√≥n y slicing b√°sico

Filas y columnas:

````python
m = np.array([[10,20,30],[40,50,60],[70,80,90]])

m[0],      # fila 0
m[:,1],    # columna 1
m[1,2]     # elemento fila 1 columna 2
````

---
## 4Ô∏è‚É£ Slicing multidimensional

Bloques:

````python
m[:2, :2], m[1:, 1:]
````

---
## 5Ô∏è‚É£ Indexaci√≥n booleana

Filtrar datos:

````python
v = np.array([10,15,22,7,5,80])
v[v > 20]
````

Tambi√©n funciona en matrices:

````python
m[m >= 50]
````

---
## 6Ô∏è‚É£ Indexaci√≥n avanzada (Fancy Indexing)

Permite seleccionar posiciones arbitrarias:

````python
v[[0,3,4]]
````

Matrices tambi√©n soportan fancy indexing por filas y columnas:

````python
m[[0,2], [1,2]]  # m[0,1] y m[2,2]
````
