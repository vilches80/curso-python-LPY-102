## 3ï¸âƒ£ IndexaciÃ³n bÃ¡sica, slicing y booleana

### ðŸ§© Ejercicio
Dada la matriz:

5. Usa fancy indexing para obtener la diagonal inversa

## 4ï¸âƒ£ Copias vs Vistas â€” Comportamiento real

### ðŸ§© Ejercicio
Usa el mismo array `m`:

1. ObtÃ©n una vista de la primera fila y modifica un valor
2. Comprueba si `m` cambia
3. ObtÃ©n una copia de esa fila y modifica un valor
4. Comprueba si `m` cambia
5. Usa `np.shares_memory` para documentar el resultado

Escribe tu soluciÃ³n aquÃ­:

````python
import numpy as np

m2 = np.arange(1,17).reshape(4,4)

print("//--- Datos ini:, "---//"", m2)

fila_vista = m2[0]
print("1. 1Âº Fila (vista):", fila_vista)

fila_vista[1] = 999
print("2. 1Âº Fila (vista) con cambio:", fila_vista)
print("2. Datos:", m2)


fila_copia = m2[0].copy()
print("3. 1Âº Fila (copia):", fila_copia)

fila_copia[0] = 999
print("4. 1Âº Fila (copia) con cambio:", fila_copia)
print("5. Datos:", m2)

print("5. Mem compartida matriz-vista:", np.shares_memory(m2, fila_vista))
print("5. Mem compartida matriz-copia:", np.shares_memory(m2, fila_copia))

print("//--- Datos fin:", m2, "---//")

````

## 5ï¸âƒ£ Ãlgebra lineal â€” ejercicios aplicados

### ðŸ§© Problema A â€” Matriz cuadrada
Dada:
```python
A = np.array([[3,1],[2,4]])
```
Calcula:
1. Determinante
2. Inversa
3. Autovalores y autovectores
4. Verifica que `A @ inv â‰ˆ I`

### ðŸ§© Problema B â€” Sistema lineal
Resuelve:
```
2x + y = 10
x + 3y = 13
```
Escribe aquÃ­ tu soluciÃ³n:

````python
import numpy as np

A = np.array([[3,1],[2,4]])

print("//--- Datos ini:", A, "---//")

detA = np.linalg.det(A)
print("1. Determinante:", detA)

invA = np.linalg.inv(A)
print("2. Inversa:", invA)

autovalA, autovecA = np.linalg.eig(A)
print("3. Autovalores:", autovalA)
print("3. Autovectores:", autovecA)

prodA = A @ invA
print("4. Interseccion:", prodA)

print("//--- Datos fin:", A, "---//")
````
