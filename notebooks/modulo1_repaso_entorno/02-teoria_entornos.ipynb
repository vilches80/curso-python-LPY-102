## 1Ô∏è‚É£ ¬øQu√© es un entorno virtual?

Un **entorno virtual** es una carpeta que contiene su propia instalaci√≥n aislada de Python.

Se usa para:
- Evitar conflictos entre versiones de librer√≠as.
- Aislar proyectos independientes.
- Garantizar reproducibilidad.
- Mantener limpio tu sistema.

üî• **Importante:** cada proyecto debe tener su propio entorno.


### ‚úîÔ∏è Crear un entorno virtual

En una terminal ejecuta:
```bash

```

### ‚úîÔ∏è Activarlo
En Linux/macOS:
```bash
source .venv/bin/activate
```
En Windows:
```bash
.venv\\Scripts\\activate
```

### ‚úîÔ∏è Ver si est√° activo
Debe aparecer algo como:

## 3Ô∏è‚É£ Estructura profesional de un proyecto

La estructura recomendada es:

````python
%%bash
# Ejemplo: crear estructura desde el notebook (solo se ejecuta si est√°s en un entorno de desarrollo real)
mkdir -p proyecto_demo/src/paquete_demo
mkdir -p proyecto_demo/tests
touch proyecto_demo/src/paquete_demo/__init__.py
echo "def suma(a,b): return a+b" > proyecto_demo/src/paquete_demo/ops.py
echo "from paquete_demo.ops import suma\n\ndef test_suma(): assert suma(2,3)==5" > proyecto_demo/tests/test_ops.py

echo "Estructura creada en proyecto_demo/"
````
